```python
import math as m

n=int(input('Number of documents:'))
l=[]
sense=[]
d=[]
dic={}

#sense maintains all senses in every doc

#dic maintains sense and corresponding words
for i in range(n):
   
    s=input('sense:')
    sense.append(s)
   
    d=input('words:').split()
    if s in dic:
        for i in d:
            dic[s].append(i)
    else:
        dic[s]=d
   
   
vocab=[]
for i in dic.values():
    for j in i:
        vocab.append(j)
       
   

#Finding prior probability
sense_prob={}
sense_set=set(sense)
for i in sense_set:
    sense_prob[i]=sense.count(i)/n
   
#Finding conditional probability


cp={}
                          #Conditional probability dictionary of all senses with sense as key
vocab_set=set(vocab)
v=len(vocab_set)
for i in sense_set:
    cps={}                           #Conditional probability of each sense with each word as key
    for w in vocab_set:
       
       
        c_ws=dic[i].count(w)
        c_s=sense.count(i)
        p_ws=(c_ws+1)/(c_s+v)
        cps[w]=p_ws
    cp[i]=cps
   

bay_dic={}
l=input("Enter Query Document:").split()
for i in sense_set:
    si=0
    for j in l:
        si+=m.log10(cp[i][j])
    si+=m.log10(sense_prob[i])
    bay_dic[i]=si

l=list(bay_dic.items())
l.sort(key=lambda x:x[1],reverse=True)
print(l[0][0])
```
