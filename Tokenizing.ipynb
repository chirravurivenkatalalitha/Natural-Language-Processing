```python
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter
s='Coal is a valuable hard, black material extracted from mines. Wood that has been buried for a long time becomes coal due to a chemical change. Earthquakes cause vast forest areas to sink underground and contribute to such changes as a result of tremendous heat and pressure. Coal mines can be found in our country at  Dhanbad, Jharia, Giridih, Chaibasa, and other locations. Coal is exported from India to Japan, Nepal, and Bangladesh. Coal is used as a fuel in both homes and factories and industries. The majority of trains and steamers move by burning coal in steam engines.'

t=word_tokenize(s)
print(t)
t=Counter(t)
for i in t:
    if i.isalnum():
        print(i,':',t[i])
        
s=s.replace(',','')
s=s.replace('.','')
        
t=word_tokenize(s)
s=' '.join(t)
j=1
l=[]         #list to hold all bigrams
for i in range(len(t)-1):
    k=t[i]+' '+t[j]
    l.append(k)
    j+=1
l=Counter(l)  #count of all bigrams
print(l)
print()
for i in l:
    l[i]=round(l[i]/len(s),3)      #probability of all bigrams
    
print(l)


```
